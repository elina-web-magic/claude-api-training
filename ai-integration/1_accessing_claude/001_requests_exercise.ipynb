# Accessing Claude with the API: Requests and Conversations

This notebook covers:
1. **Making a request** to Claude using the Anthropic API.
2. **Multi-turn conversations** using the `messages` array.
3. **Chat exercise**: building an interactive command-line chatbot.

```python
# Load env variables
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
# Helpers to write history
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=100,
        messages=messages
    )
    return message.content[0].text
```

## 1. Making a Request & Multi-Turn Conversations

```python
# Start a conversation history
messages = []

# 1. First Turn
add_user_message(messages, "what's 1+1?")
answer = chat(messages)
print("User: what's 1+1?")
print("Claude:", answer)
add_assistant_message(messages, answer)

print("-" * 20)

# 2. Second Turn (context-aware)
add_user_message(messages, "add 1 more")
answer = chat(messages)
print("User: add 1 more")
print("Claude:", answer)
add_assistant_message(messages, answer)
```

## 2. Chat Exercise (Interactive Chatbot Loop)

```python
# Starting a fresh list of messages for the chatbot
messages = []
print("Chatbot started! Type 'exit' or press Enter on an empty line (or press Esc) to quit.\n")

while True:
    try:
        # Get user input
        user_input = input("> ").strip()
        print(f"> {user_input}")
    except (KeyboardInterrupt, EOFError):
        # Catch Esc key or interrupt gracefully in environments that raise exceptions
        print("\nExiting chat. Goodbye!")
        break

    # If the user presses Esc or cancels, VS Code may return an empty string.
    # Exiting on empty input prevents getting trapped in an infinite prompt loop!
    if not user_input:
        print("Exiting chat. Goodbye!")
        break

    # Check if the user wants to exit via text command
    if user_input.lower() in ["exit", "quit"]:
        print("Exiting chat. Goodbye!")
        break

    # Add user input to history
    add_user_message(messages, user_input)
    
    # Call Claude
    answer = chat(messages)
    
    # Add generated text to history
    add_assistant_message(messages, answer)

    # Print response
    print("---")
    print(answer)
    print("---")
```
