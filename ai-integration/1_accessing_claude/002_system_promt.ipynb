# Accessing Claude with the API: System Prompts

This notebook covers:
1. **System Prompts**: Instructing and framing Claude's behavior using the `system` parameter.
2. **System Prompts Exercise**: Creating specialized personas (e.g., highly concise developer profile).

```python
# Load env variables
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
# Helpers to write history
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 600,
        "messages": messages
    }
    if system:
        params["system"] = system
        
    message = client.messages.create(**params)
    return message.content[0].text
```

## System Prompts Exercise

```python
from IPython.display import display, Markdown

messages = []

# Define the system instruction to shape Claude's role/personality
system_instruction = "You are a Python engineer who writes very concise code."

add_user_message(messages, "Generate a one sentence movie idea")
answer = chat(messages, system=system_instruction)

display(Markdown(answer))
```
