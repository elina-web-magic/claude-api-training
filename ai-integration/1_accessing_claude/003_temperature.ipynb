# Accessing Claude with the API: Temperature

This notebook covers:
1. **Temperature**: Modifying the model's output randomness to control creativity vs determinism.

```python
# Load env variables
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
# Helpers to write history
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, temperature=1.0):
    message = client.messages.create(
        model=model,
        max_tokens=600,
        messages=messages,
        temperature=temperature
    )
    return message.content[0].text
```

## Controlling Determinism & Creativity with Temperature

- **Temperature 1.0 (High)**: More creative, unpredictable, and diverse outputs.
- **Temperature 0.1 (Low)**: More deterministic, standard, and consistent outputs.

```python
# 1. High Temperature (Creative)
print("=== Temperature 1.0 (Creative Runs) ===")
for i in range(3):
    messages = []
    add_user_message(messages, "Write a one-sentence catchphrase for a dog food brand")
    print(f"Run {i+1}: {chat(messages, temperature=1.0)}")

print("\n" + "=" * 40 + "\n")

# 2. Low Temperature (Deterministic)
print("=== Temperature 0.1 (Deterministic Runs) ===")
for i in range(3):
    messages = []
    add_user_message(messages, "Write a one-sentence catchphrase for a dog food brand")
    print(f"Run {i+1}: {chat(messages, temperature=0.1)}")
```
